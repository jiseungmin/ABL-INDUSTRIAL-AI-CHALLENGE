```python
# Bad
#        erode
#         . dilate

imgPath = "/Users/jiseungmin/Desktop/chunnamAIcontest/livetdata_bad/"
outputPath = "/Users/jiseungmin/Desktop/chunnamAIcontest/livetdata_bad_result/"
file_lst = os.listdir(imgPath)

def binaryDef(imgPath) :
  k = cv2.getStructuringElement(cv2.MORPH_RECT, (2,2))
  src = cv2.imread(imgPath, cv2.IMREAD_COLOR)
  gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
  ret, dst = cv2.threshold(gray, 20, 255, cv2.THRESH_BINARY)
  dst = cv2.morphologyEx(dst, cv2.MORPH_CLOSE, k, iterations=2)

  
  img_save(dst)

def img_save(dst):
  cv2.imwrite(imgPath+"result.png", dst)
  background = Image.open("/Users/jiseungmin/Desktop/chunnamAIcontest/image/background.jpeg")
  background = background.resize((256,256))
  foreground = Image.open(imgPath+"result.png")
  background.paste(foreground, (0, 0), foreground)
  background.save(outputPath+file)


for file in file_lst :
	binaryDef(imgPath+file)
```

```python
# Pass
#        erode
#         . dilate

imgPath = "/Users/jiseungmin/Desktop/chunnamAIcontest/livetdata_pass/"
outputPath = "/Users/jiseungmin/Desktop/chunnamAIcontest/livetdata_pass_result/"
file_lst = os.listdir(imgPath)

def binaryDef(imgPath) :
  k = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
  src = cv2.imread(imgPath, cv2.IMREAD_COLOR)
  gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
  ret, dst = cv2.threshold(gray, 15, 255, cv2.THRESH_BINARY)
  dst = cv2.dilate(dst,k,iterations=4)
  img_save(dst)

def img_save(dst):
  cv2.imwrite(imgPath+"result.png", dst)
  background = Image.open("/Users/jiseungmin/Desktop/chunnamAIcontest/image/background.jpeg")
  background = background.resize((256,256))
  foreground = Image.open(imgPath+"result.png")
  background.paste(foreground, (0, 0), foreground)
  background.save(outputPath+file)


for file in file_lst :
	binaryDef(imgPath+file)
```
